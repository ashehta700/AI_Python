```python
import random

def main():
    choices = ["Rock", "Paper", "Scissor"]

    print("Welcome to Rock Paper Scissor Game!")
    print("Note: You are playing against the computer.")

    name = input("Please enter your name: ")

    play_again = "yes"

    while play_again.lower() == "yes":
        print("""
Please choose one of the options:
1. Rock
2. Paper
3. Scissor
        """)

        user_choice = int(input("Enter your choice (1-3): "))
        user_choice = choices[user_choice - 1]

        computer_choice = random.choice(choices)

        print(f"{name} chose {user_choice}")
        print(f"Computer chose {computer_choice}")

        if user_choice == computer_choice:
            print("it's fair No one win ")
        elif (
            (user_choice == "Rock" and computer_choice == "Scissor") or
            (user_choice == "Paper" and computer_choice == "Rock") or
            (user_choice == "Scissor" and computer_choice == "Paper")
        ):
            print(f"Congratulations {name}! You win!")
        else:
            print(f"Sorry {name}! Computer wins!")

        play_again = input("Do you want to play again? (yes/no): ")

    print("Thank you for playing!")

main()
```

```python


# task 2
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error: Division by zero is not allowed."

def multiply(x, y):
    return x * y

def is_even(x):
    return x % 2 == 0

def is_odd(x):
    return x % 2 != 0

def factorial(x):
    if x < 0:
        return "Error: Factorial is not defined for negative numbers."
    elif x == 0 or x == 1:
        return 1
    else:
        result = 1
        for i in range(2, x + 1):
            result *= i
        return result

def fibonacci(n):
    # this the method
    # Fn = Fn-1+Fn-2
    if n < 0:
        return "Error: Fibonacci sequence is not defined for negative numbers."
    elif n == 0:
        return []
    elif n == 1:
        return [0]
    else:
        sequence = [0, 1]
        while len(sequence) < n:
            next_num = sequence[-1] + sequence[-2]
            sequence.append(next_num)
        return sequence


def main():
    print("""
Welcome to Calculator App!")
Please choose one of the following operations:")
1. Add
2. Subtract
3. Divide
4. Multiply
5. Check if a number is even
6. Check if a number is odd
7. Factorial
8. Fibonacci Sequence""")

    choice = int(input("Enter your choice (1-8): "))

    if choice in [1, 2, 3, 4]:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        if choice == 1:
            result = add(num1, num2)
            print("Result:", result)
        elif choice == 2:
            result = subtract(num1, num2)
            print("Result:", result)
        elif choice == 3:
            result = divide(num1, num2)
            print("Result:", result)
        elif choice == 4:
            result = multiply(num1, num2)
            print("Result:", result)
    elif choice in [5, 6]:
        num = int(input("Enter the number: "))

        if choice == 5:
            result = is_even(num)
            print(f"{num} is even:", result)
        elif choice == 6:
            result = is_odd(num)
            print(f"{num} is odd:", result)
    elif choice == 7:
        num = int(input("Enter the number: "))
        result = factorial(num)
        print(f"Factorial of {num}:", result)
    elif choice == 8:
        num = int(input("Enter the number of terms: "))
        result = fibonacci(num)
        print("Fibonacci Sequence:", result)
    else:
        print("Invalid choice. Please choose a number between 1 and 8.")

    print("Thank you for using the Calculator App!")

main()
```
